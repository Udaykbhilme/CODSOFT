```python
# Drop the 'Age' column from the DataFrame.
df= df.drop(['Age'], axis=1)
```

```python
# Create a copy of the DataFrame and assign it to df_final.
df_final =df
# Display the first 10 rows of the df_final DataFrame.
df_final.head(10)
```

```python
# Select 'Pclass' and 'Sex' columns as features for the model.
x = df[['Pclass','Sex']]
# Select 'Survived' as the target variable for the model.
y = df['Survived']
```

```python
# Split the data into training and testing sets.
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state =0)
```

```python
# Import and instantiate the Logistic Regression model.
from sklearn.linear_model import LogisticRegression
log = LogisticRegression(random_state=0)
log.fit(x_train, y_train)
```

 Prediction

```python
# Make predictions on the test data and print the results.
pred = print(log.predict(x_test))
```

```python
# Print the actual values of the test set target variable.
print(y_test)
```

```python
# Print "Not Survived" or "Survived" based on the prediction result.
import warnings
warnings.filterwarnings('ignore')

res = log.predict([[2,0]])

if (res == 0):
    print("Not Survived")
else:
    print("Survived")
```
